```python
#manually selection 

import json
import numpy as np
import matplotlib.pyplot as plt
import os
counter = 0
Path = '/content.gdrive/MyDrive/Vorselektion/Landing_sequence_big/Results'
files = next(os.walk(Path))[1]
for s in files:
  runs = next(os.walk(Path + '/' + s))[1]
  for r in runs: 
    print(s + r)   
    with open(Path + '/' + s + '/' + r + '/list.json', "r") as fp:
        data = json.load(fp)
      
    conf = np.array(data["conf"])
    index = np.array(data["index"])
    cls = np.array(data["cls"])
    x_center = np.array(data["x"])
    y_center = np.array(data["y"])
    if index[0] == 0:
      break





    index1 = []
    conf1 = []
    cls1 = []
    x_center1 = []
    y_center1 = []
    index2 = []
    conf2 = []
    cls2 = []
    x_center2 = []
    y_center2 = []
    index3 = []
    conf3 = []
    cls3 = []
    x_center3 = []
    y_center3 = []
    index4 = []
    conf4 = []
    cls4 = []
    x_center4 = []
    y_center4 = []


   
    for i, idx in enumerate(index):
            if idx[0] == '1':
              conf1.append(conf[i])
              index1.append(idx[2:])
              cls1.append(cls[i])
            elif idx[0] == '2':
              conf2.append(conf[i])
              index2.append(idx[2:])
              cls2.append(cls[i])
            elif idx[0] == '3':
              conf3.append(conf[i])
              index3.append(idx[2:])
              cls3.append(cls[i])
            elif idx[0] == '4':
              conf4.append(conf[i])
              index4.append(idx[2:])
              cls4.append(cls[i])


    detections = []
    cams = [conf1, conf2, conf3, conf4]
    indexes = [index1, index2, index3, index4]
    for idx, cam in enumerate(cams):
              for i, c in enumerate(cam):
                if c > 0:
                  detections.append(str(idx+1)+ "_" + indexes[idx][i])
    #print(detections)
    for i, d in enumerate(detections):
        if d[0] == '1':
          x_center1.append(x_center[i])
          y_center1.append(y_center[i])
        elif d[0] == '2':
          x_center2.append(x_center[i])
          y_center2.append(y_center[i])            
        elif d[0] == '3':
          x_center3.append(x_center[i])
          y_center3.append(y_center[i])
        elif d[0] == '4':
          x_center4.append(x_center[i])
          y_center4.append(y_center[i])



    print(x_center1)
    plt.plot(x_center1,y_center1,label='Cam1')
    plt.xlim(0, 1080) # set x-axis limits to (0, 1080)
    plt.ylim(0, 1920) # set y-axis limits to (0, 1920)
    plt.gca().invert_yaxis()
    #plt.show()
    plt.plot(x_center2,y_center2,label='Cam2')
    plt.xlim(0, 1080) # set x-axis limits to (0, 1080)
    plt.ylim(0, 1920) # set y-axis limits to (0, 1920)
    plt.gca().invert_yaxis()
    #plt.show()
    plt.plot(x_center3,y_center3,label='Cam3')
    plt.xlim(0, 1080) # set x-axis limits to (0, 1080)
    plt.ylim(0, 1920) # set y-axis limits to (0, 1920)
    plt.gca().invert_yaxis()
    #plt.show()
    plt.plot(x_center4,y_center4,label='Cam4')
    plt.xlim(0, 1080) # set x-axis limits to (0, 1080)
    plt.ylim(0, 1920) # set y-axis limits to (0, 1920)
    plt.gca().invert_yaxis()
    plt.legend(loc='upper left')
    plt.show()
    

    counter += 1
    if counter == 15:

      x = 1/0








```
