```python
from tensorflow import keras
import cv2
import numpy as np
import os
from PIL import Image

resnet_model = keras.models.load_model('/content.gdrive/MyDrive/Data_Resnet_VAL/resnet_big_100_Epochs.h5')
class_names = ['FLight', 'Landing']
```

```python
from PIL import Image
#Path_Jump = '/content.gdrive/MyDrive/Vorselektion/B-Hofen_Samples/30'
PATH = '/content.gdrive/MyDrive/Resnet_Test/' + str(Path_Jump[-2:])
pics = []
try:
    pics = next(os.walk(PATH))[2]
except StopIteration:
    pass



for pic in pics:

# √ñffnen Sie das Bild
    im = Image.open(PATH + '/' + str(pic))

    if im.width < 400 or im.height < 400:
      # Vergr√∂√üern Sie das Bild auf 400x400
      im_resized = im.resize((400, 400), resample=Image.NEAREST)

      # Erstellen Sie eine neue schwarze Fl√§che, die die Gr√∂√üe des vergr√∂√üerten Bildes hat
      im_black = Image.new("RGB", (im_resized.width, im_resized.height), (255, 255, 255))

      # F√ºgen Sie das vergr√∂√üerte Bild in die Mitte der schwarzen Fl√§che ein
      im_black.paste(im_resized, (int((im_black.width - im_resized.width) / 2), int((im_black.height - im_resized.height) / 2)))

      # Speichern Sie das neue Bild
      im_black.save(PATH + '/' + str(pic))




    image=cv2.imread(PATH + '/' + str(pic))
    image=np.expand_dims(image,axis=0)

    pred=resnet_model.predict(image)
    print(pred)
        


    output_class=class_names[np.int32(np.around(pred[0,0]))]
    print("The predicted class for ",str(pic), "is", output_class)
    

```
