```python
#danach durchgehen und reahpen 400x400
from PIL import Image
import os
PATH = '/content.gdrive/MyDrive/Data_Resnet_VAL/Landing'
label= []
try:
    label = next(os.walk(PATH))[2]
except StopIteration:
    pass
#data["labels"].append(label)
print(label)


for image in label:

# Ã–ffnen Sie das Bild
    im = Image.open("/content.gdrive/MyDrive/Data_Resnet_VAL/Landing/" + str(image))





    if im.width < 400 or im.height < 400:
      # VergrÃ¶ÃŸern Sie das Bild um den Faktor 2
      im_resized = im.resize((400, 400), resample=Image.NEAREST)

      # Erstellen Sie eine neue schwarze FlÃ¤che, die die GrÃ¶ÃŸe des vergrÃ¶ÃŸerten Bildes hat
      im_black = Image.new("RGB", (im_resized.width, im_resized.height), (255, 255, 255))

      # FÃ¼gen Sie das vergrÃ¶ÃŸerte Bild in die Mitte der schwarzen FlÃ¤che ein
      im_black.paste(im_resized, (int((im_black.width - im_resized.width) / 2), int((im_black.height - im_resized.height) / 2)))

      # Speichern Sie das neue Bild
      im_black.save("/content.gdrive/MyDrive/Data_Resnet_VAL/Landing/" + str(image))
      print("/content.gdrive/MyDrive/Data_Resnet_VAL/Landing/" + str(image))
      
    else:
      print("OK")
      
```
